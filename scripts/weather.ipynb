```python
from dataloader import DataLoader
import pandas as pd

files = ["20170" + str(i) + "-citibike-tripdata.csv" if i < 10 else "2017" + str(i) + "-citibike-tripdata.csv" for 
i in range(1,13)]

for file in files:
    dl = DataLoader("../data", [file])
    
    dl.load_data()
    df = dl.data.copy()
    df['start time'] = pd.to_datetime(df['start time'])
    if file == "20170" + str(1) + "-citibike-tripdata.csv":
        count_by_day_in_month = df.groupby([df['start time'].dt.date]).size().reset_index()
        count_by_day_in_month.columns = ['day','count']
        count_by_day_in_month.head()
    else:
        tmp = df.groupby([df['start time'].dt.date]).size().reset_index()
        tmp.columns = ['day','count']
        count_by_day_in_month = pd.concat([count_by_day_in_month,tmp])
    print(file)

count_by_day_in_month.to_csv("../data/to_app/count_by_day2017.csv",
                            index = False)
```

## wczytanie ściągniętych danych:

```python
data = pd.read_csv("../data/weather2017/201701.csv",skip_blank_lines=True
            ,skiprows=4)

def weather_data(data):
    colnames = ['Date','Maximum','Minimum','Average','Departure',
           'HDD','CDD','Precipitation','New Snow','Snow Depth']


    data.columns = colnames

    ind = []
    res = []
    for i in range(data.shape[0]):
        if data['Date'][i] == 'Sum':
            res.append(1)

        if len(res) == 0:
            ind.append(i)

    data = data.iloc[ind,:]

    
    return data
```

```python
data = pd.read_csv("../data/weather2017/201701.csv",skip_blank_lines=True
            ,skiprows=4)

res = weather_data(data)
for i in range(2,8):
    file = "../data/weather2017/20170" + str(i) if i < 10 else "../data/weather2017/2017" + str(i)
    
    if i <= 7:
        data = pd.read_csv(file + ".csv",
                           skip_blank_lines=True,skiprows=4)
    data = weather_data(data)
    res = pd.concat([res,data])
    
for column in res.columns.values[1:4]:
    res[column] = (res[column].astype(float) - 32) * (5/9)
    

res = res.iloc[:,[0,1,2,3,7,8,9]]
res.columns = weather.columns
res['date'] = pd.to_datetime(res['date'])
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
spotify_green = '#1DB954'

weather = pd.read_csv("../data/to_app/weather_data_1617.csv")
count_by_day2017 = pd.read_csv("../data/to_app/count_by_day2017.csv")
count_by_day2016 = pd.read_csv("../data/to_app/count_by_day2016.csv")
count_by_day = pd.concat([count_by_day2016,count_by_day2017]).sort_values(['day']).reset_index(drop = True)



fig = make_subplots(
    rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.02,
    print_grid=False,specs=[[{"secondary_y": False}],[{"secondary_y": True}]]
)

fig.add_trace(
    go.Scatter(x=weather.loc[:,'date'],
               y=weather.loc[:,'average temperature'].values,
               name = "temperature", marker = dict(color = spotify_green)
    ),
    row=2, col=1,secondary_y=False
)

fig.add_trace(
    go.Bar(
        x=weather.loc[:,'date'],
        y=weather.loc[:,'snow depth'].values,
        name = 'snow',marker = dict(color = "white")
    ),
    row = 2, col = 1,secondary_y=True
)

fig.add_trace(
    go.Bar(
        x=weather.loc[:,'date'],
        y=weather.loc[:,'precipitation'].values,
        name = 'rain',marker = dict(color = "blue")
    ),
    row = 2, col = 1,secondary_y=True
)



fig.add_trace(go.Scatter(x=count_by_day.loc[:,'day'].values,
                         y=count_by_day.loc[:,'count'].values,
                         name = "rides count")
              ,row=1, col=1)


fig.update_layout(height=600, width=800,
                  title_text="weather vs bike rides"
)


fig['layout']['yaxis2'].update(showgrid=False,zeroline=False)
fig['layout']['yaxis1'].update(showgrid=False,zeroline=False)
fig['layout']['xaxis2'].update(showgrid=False,zeroline=False)
fig['layout']['xaxis1'].update(showgrid=False,zeroline=False)
fig['layout']['yaxis3'].update(showgrid=False,zeroline=False)
fig.layout.plot_bgcolor = '#1E1E1E'
fig.layout.paper_bgcolor = '#1E1E1E'

fig.show()

```
