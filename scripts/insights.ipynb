```python
from dataloader import DataLoader
import pandas as pd

dl = DataLoader("../data",
                ["201701-citibike-tripdata.csv",
                "201702-citibike-tripdata.csv",
                "201703-citibike-tripdata.csv",
                "201704-citibike-tripdata.csv",
                "201705-citibike-tripdata.csv",
                "201706-citibike-tripdata.csv"])

dl.load_data()
stations = dl.load_stations()
station_counts = dl.load_station_counts()
```

```python
df = dl.data.copy()
```

```python
df['start time'] = pd.to_datetime(df['start time'])
df.head()
```

```python
df['weekday'] = df['start time'].dt.weekday_name
```

```python
df.head()
```

```python
count_by_day_in_month = df.groupby([df['start time'].dt.date,df.gender]).size().reset_index()
count_by_day_in_month.columns = ['day','gender','count']
count_by_day_in_month.head()
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=count_by_day_in_month.loc[count_by_day_in_month.gender == 1,'day'].values,
                         y=count_by_day_in_month.loc[count_by_day_in_month.gender == 1,'count'].values,
                    mode='lines',
                    name='woman'))

fig.add_trace(go.Scatter(x=count_by_day_in_month.loc[count_by_day_in_month.gender == 2,'day'].values,
                         y=count_by_day_in_month.loc[count_by_day_in_month.gender == 2,'count'].values,
                    mode='lines',
                    name='man'))
```
